```python
  import os
  import matplotlib.pyplot as plt
  from mtcnn import MTCNN
  def draw_faces(filename, result_list):
    data=plt.imread(filename)
    for i in range(len(result_list)):
      x1, y1, width, height = result_list[i]['box']
      x2, y2 = x1+width, y1+height
      plt.subplot(1, len(result_list),i+1)
      plt.axis('off')
      plt.imshow(data[y1:y2,x1:x2])
      face = data[y1:y2, x1:x2]
      plt.imshow(face)
      # create the 'faces' directory if it does not exist
      if not os.path.exists('faces'):
            os.makedirs('faces')
        # save the face image to the 'faces' directory
      plt.imsave('faces/face{}.jpg'.format(i+1), face)
    plt.show()
  filename='/content/drive/MyDrive/Anh/marifactory297844494627012510-20171027163016.jpg'
  img=plt.imread(filename)
  detector=MTCNN()
  faces=detector.detect_faces(img)
  draw_faces(filename,faces)
```
